```python
from PIL import Image
import urllib
import numpy as np

URL = 'http://ecx.images-amazon.com/images/I/41qJGxrMW0L._SL500_AA300_.jpg'
urllib.urlretrieve(URL, "image.jpg")
```

```python
im = Image.open('image.jpg', 'r')
width, height = im.size
pixel_values = list(im.getdata())

print(width)
print(height)
print(pixel_values)
```
