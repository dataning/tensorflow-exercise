```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import numpy as np


from six.moves import urllib

import matplotlib.pyplot as plt
%matplotlib inline 

import tensorflow as tf 
```

```python
filenames = ['data/san_francisco.jpg']
filename_queue = tf.train.string_input_producer(filenames)

print(filename_queue)

reader = tf.WholeFileReader()
key, value = reader.read(filename_queue)

images = tf.image.decode_jpeg(value, channels=3)

#print(images)

init_op = tf.initialize_all_variables()
with tf.Session() as sess:
  sess.run(init_op)
  
  coord = tf.train.Coordinator()
  threads = tf.train.start_queue_runners(coord=coord)

  for i in range(1): #length of your filename list
      image = images.eval() #here is your image Tensor :) 

  print(image.shape)

  pix = np.array(image)
  print(pix)
#  Image.show(Image.fromarray(np.asarray(image)))

plt.imshow(pix)
```

```python
rand = tf.constant(
    [[[1, 1, 1], [2, 2, 2]],
     [[3, 3, 3], [4, 4, 4]],
     [[5, 5, 5], [6, 6, 6]]])

slice = tf.slice(rand,[1,0,0],[1,1,3])
slice2 = tf.slice(rand,[1,0,0],[1,2,3])
slice3 = tf.slice(rand,[1,0,0],[2,1,3])
slice4 = tf.slice(rand,[0,0,0],[2,2,3])

print(rand.get_shape())

sess = tf.Session()
result = sess.run(slice)
print('')
print(result)

result2 = sess.run(slice2)
print('')
print(result2)

result3 = sess.run(slice3)
print('')
print(result3)

result4 = sess.run(slice4)
print('')
print(result4)
```

```python
import math

embedding_size = 128
stddev=1.0 / math.sqrt(embedding_size)
print(stddev)
```

```python
import tensorflow as tf

vocabulary_size = 50000
picture = tf.truncated_normal([vocabulary_size, embedding_size],
                            stddev=1.0 / math.sqrt(embedding_size))

sess = tf.Session()
result = sess.run(picture)
print('')
print(result)
print(type(result))
print(result[0,1])

print(result[0,:].size)

xline = np.arange(0,128,1)


plt.scatter(xline, result[0,:])
plt.show()
```
