```python
import numpy as np
import math
import tensorflow as tf
```

```python


a= np.ndarray(shape=(8), dtype=np.int32)
b= np.ndarray(shape=(8,1), dtype=np.int32)

print(a)
print(b)

print(8//5)

print(1.0 / math.sqrt(128))
```

- tensorflow tensor

```python
test1 = tf.truncated_normal([3, 3], mean=0.0, stddev=0.02, dtype=tf.float32, seed=1)
sess = tf.Session()
print(sess.run(test1))

sess.close()



x = tf.placeholder(shape=[2], dtype=tf.float32)
y = tf.inv(x)
sess = tf.Session()
rand_array = np.random.rand(2)
print(rand_array)
print(sess.run(y, feed_dict={x: rand_array})) 

sess.close()
```

```python
# mean, std 

sess = tf.Session()
rand_var = tf.random_normal([5],0,10, dtype = tf.float32, seed = 0)
print(sess.run(rand_var))
sess.close()

# min value, max value
sess = tf.Session()
rand_var = tf.random_uniform([5],0,10, dtype = tf.float32, seed = 0)
print(sess.run(rand_var))
sess.close()

# numpy

print(np.random.random((8, 3)))
print(1.0 / math.sqrt(128))
```

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

%pylab inline

iris = load_iris()
X_tsne = TSNE(learning_rate=100).fit_transform(iris.data)
X_pca = PCA().fit_transform(iris.data)
```

```python
figure(figsize=(10, 5))
subplot(121)
scatter(X_tsne[:, 0], X_tsne[:, 1], c=iris.target)
subplot(122)
scatter(X_pca[:, 0], X_pca[:, 1], c=iris.target)
```
